```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel 파일에서 데이터 읽기
df = pd.read_excel("./의사수간호사수_병상수.xlsx")
df = df.sort_values(by="지역")
# 데이터 설정
x = df['지역']
bar_data1 = df['2020의사']
bar_data2 = df['2020간호사']
line_data = df['20_인구천명당병상개수']

fig, ax1 = plt.subplots(figsize=(15, 6))
plt.xticks(rotation=45)
# 두 개의 막대 그래프를 나란히 그립니다.
width = 0.4
x_indexes = np.arange(len(x))
ax1.bar(x_indexes - width/2, bar_data1, width, color='#656ADC', alpha=0.7, label='2020의사')
ax1.bar(x_indexes + width/2, bar_data2, width, color='#67C88E', alpha=0.7, label='2020간호사')

# 보조 선 그래프 추가
ax2 = ax1.twinx()
ax2.plot(x_indexes, line_data, 'r--', marker='o', label='인구천명당 병상개수')


# 각 막대 바 위에 값을 표시
for i, (val1, val2) in enumerate(zip(bar_data1, bar_data2)):
    ax1.text(x_indexes[i] - width/2, val1, str(val1), ha='center', va='bottom')
    ax1.text(x_indexes[i] + width/2, val2, str(val2), ha='center', va='bottom')

# 각 점 위에 값을 표시
for i, y_val in enumerate(line_data):
    ax2.text(x_indexes[i], y_val, str(y_val), ha='right', va='bottom')

# X축 레이블 설정
ax1.set_xticks(x_indexes)
ax1.set_xticklabels(x)

# 축 레이블 및 범례 추가
ax1.set_xlabel("지역")


# 그래프 제목
plt.title("병상수에 따른 의사, 간호사 수")

# 범례 표시
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 그래프 표시
plt.tight_layout()
plt.show()
```
